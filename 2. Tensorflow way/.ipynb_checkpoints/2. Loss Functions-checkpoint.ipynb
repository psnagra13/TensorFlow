# Loss Functions

```python
import matplotlib.pyplot as plt
import tensorflow as tf
sess = tf.Session()
```

## Loss Function for Linear Regression

```python
x_vals = tf.linspace(-1., 1., 500)
target = tf.constant(0.)
```

L2 norm loss is also known as the Euclidean loss function.

It is just the square of the distance to the target.

```python
l2_y_vals = tf.square(target - x_vals)
l2_y_out = sess.run(l2_y_vals)
# Inbuilt method : nn.l2_loss()
```

L1 norm loss is also known as the absolute loss function. 

Instead of squaring the difference, we take the absolute value.

```python
l1_y_vals = tf.abs(target - x_vals)
l1_y_out = sess.run(l1_y_vals)
```

## Loss Function for Classification

```python
x_vals = tf.linspace(-3., 5., 500)
target = tf.constant(1.)
targets = tf.fill([500,], 1.)
```

<b> Hinge Loss </b> : It is meant to compute a loss between with two target classes, 1 and -1.

Hinge loss is mostly used for support vector machines, but can be used in neural networks as well.

```python
hinge_y_vals = tf.maximum(0., 1. - tf.multiply(target, x_vals))
hinge_y_out = sess.run(hinge_y_vals)
```

<b>Cross-entropy loss </b> for a binary case is also sometimes referred to as the logistic loss function. 

It comes about when we are predicting the two classes 0 or 1.

```python
xentropy_y_vals = - tf.multiply(target, tf.log(x_vals)) - tf.multiply((1. - target), tf.log(1. - x_vals))
xentropy_y_out = sess.run(xentropy_y_vals)
```

<b>Sigmoid cross entropy loss </b> is very similar to the previous loss function except we transform the x-values by the sigmoid function
before we put them in the cross entropy loss

```python
xentropy_sigmoid_y_vals = tf.nn.sigmoid_cross_entropy_with_logits(logits=x_vals, labels=targets)
xentropy_sigmoid_y_out = sess.run(xentropy_sigmoid_y_vals)
```

<b>Softmax cross-entropy loss </b> operates on non-normalized outputs. 

This function is used to measure a loss when there is only one target category instead of multiple.

```python
unscaled_logits = tf.constant([[1., -3., 10.]])
target_dist = tf.constant([[0.1, 0.02, 0.88]])
softmax_xentropy = tf.nn.softmax_cross_entropy_with_logits(logits=unscaled_logits, labels= target_dist)
print(sess.run(softmax_xentropy))
```

## Graphical Visualization of Regression loss functions

```python
x_array = sess.run(x_vals)
plt.plot(x_array, l2_y_out, 'b-', label='L2 Loss')
plt.plot(x_array, l1_y_out, 'r--', label='L1 Loss')
plt.ylim(-0.2, 0.4)
plt.legend(loc='lower right', prop={'size': 11})
plt.show()
```

## Graphical Visualization of Classification loss functions

```python
x_array = sess.run(x_vals)
plt.plot(x_array, hinge_y_out, 'b-', label='Hinge Loss')
plt.plot(x_array, xentropy_y_out, 'r--', label='Cross EntropyLoss')
plt.plot(x_array, xentropy_sigmoid_y_out, 'k-.', label='CrossEntropy Sigmoid Loss')
plt.ylim(-1.5, 3)
plt.legend(loc='lower right', prop={'size': 11})
plt.show()
```
